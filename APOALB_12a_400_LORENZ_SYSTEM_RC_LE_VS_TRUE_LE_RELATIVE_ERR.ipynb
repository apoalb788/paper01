# Standard Reservoir Computer

### absolute and relative error estimation of Lyapunov exponents calculated with RC
### (compared to those calculated of the true system)

```python
# Import libraries
import math     as m
import numpy    as np
import matplotlib.pyplot as plt  
import pandas   as pd
import time     as time

#######################################################################
# E N V I R O N M E N T   S E T   U P
#######################################################################
#---------------------------------------------------------------------#
# To compute elapsed time
#---------------------------------------------------------------------#
start_time = time.time()
```

```python
#######################################################################
# T R U E   S Y S T E M    L Y A P U N O V   E X P O N E N T S   L I S T
#######################################################################
df_True_LE = pd.read_csv('climate/df_True_LE_x_PAPER.csv',delimiter=";",index_col=False)

len_df_True_LE = len(df_True_LE)
```

```python
D = 3

#######################################################################
# R. C .  L Y A P U N O V   E X P O N E N T S   L I S T
#######################################################################
N = 400
txt_seed = 'seed42' 
df_RC_LE_seed42 = pd.read_csv('climate/df_RC' + str(N) + '_' + txt_seed + '_LE_PAPER.csv',delimiter=";",index_col=False)

txt_seed = 'seed84' 
df_RC_LE_seed84 = pd.read_csv('climate/df_RC' + str(N) + '_' + txt_seed + '_LE_PAPER.csv',delimiter=";",index_col=False)

txt_seed = 'seed167' 
df_RC_LE_seed167 = pd.read_csv('climate/df_RC' + str(N) + '_' + txt_seed + '_LE_PAPER.csv',delimiter=";",index_col=False)

len_df_RC_LE = len(df_RC_LE_seed42)
rho_list = df_RC_LE_seed42.rho_Lorenz
```

```python
df_RC_LE_seed42['abs_err_1'] = np.abs(df_RC_LE_seed42.lambda_1 - df_True_LE.lambda_1)
df_RC_LE_seed42['abs_err_2'] = np.abs(df_RC_LE_seed42.lambda_2 - df_True_LE.lambda_2)
df_RC_LE_seed42['abs_err_3'] = np.abs(df_RC_LE_seed42.lambda_3 - df_True_LE.lambda_3)

df_RC_LE_seed42['rel_err_1'] = np.abs((df_RC_LE_seed42.lambda_1 - df_True_LE.lambda_1)/df_True_LE.lambda_1)
df_RC_LE_seed42['rel_err_2'] = np.abs((df_RC_LE_seed42.lambda_2 - df_True_LE.lambda_2)/df_True_LE.lambda_2)
df_RC_LE_seed42['rel_err_3'] = np.abs((df_RC_LE_seed42.lambda_3 - df_True_LE.lambda_3)/df_True_LE.lambda_3)
               
df_RC_LE_seed84['abs_err_1'] = np.abs(df_RC_LE_seed84.lambda_1 - df_True_LE.lambda_1)
df_RC_LE_seed84['abs_err_2'] = np.abs(df_RC_LE_seed84.lambda_2 - df_True_LE.lambda_2)
df_RC_LE_seed84['abs_err_3'] = np.abs(df_RC_LE_seed84.lambda_3 - df_True_LE.lambda_3)

df_RC_LE_seed84['rel_err_1'] = np.abs((df_RC_LE_seed84.lambda_1 - df_True_LE.lambda_1)/df_True_LE.lambda_1)
df_RC_LE_seed84['rel_err_2'] = np.abs((df_RC_LE_seed84.lambda_2 - df_True_LE.lambda_2)/df_True_LE.lambda_2)
df_RC_LE_seed84['rel_err_3'] = np.abs((df_RC_LE_seed84.lambda_3 - df_True_LE.lambda_3)/df_True_LE.lambda_3)

df_RC_LE_seed167['abs_err_1'] = np.abs(df_RC_LE_seed167.lambda_1 - df_True_LE.lambda_1)
df_RC_LE_seed167['abs_err_2'] = np.abs(df_RC_LE_seed167.lambda_2 - df_True_LE.lambda_2)
df_RC_LE_seed167['abs_err_3'] = np.abs(df_RC_LE_seed167.lambda_3 - df_True_LE.lambda_3)

df_RC_LE_seed167['rel_err_1'] = np.abs((df_RC_LE_seed167.lambda_1 - df_True_LE.lambda_1)/df_True_LE.lambda_1)
df_RC_LE_seed167['rel_err_2'] = np.abs((df_RC_LE_seed167.lambda_2 - df_True_LE.lambda_2)/df_True_LE.lambda_2)
df_RC_LE_seed167['rel_err_3'] = np.abs((df_RC_LE_seed167.lambda_3 - df_True_LE.lambda_3)/df_True_LE.lambda_3)
```

```python
for i in range(len_df_True_LE):
    print(rho_list[i],df_RC_LE_seed42.rel_err_1[i], df_RC_LE_seed42.rel_err_2[i], df_RC_LE_seed42.rel_err_3[i])
```

```python
# Plot Lorenz system Lyapunov Exponents
fig, axs = plt.subplots(2, 1, figsize=(8, 8))  

axis_test_label_seed42   = ["$\lambda_1$ seed=42"  ,"$\lambda_2$ seed=42"  ,"$\lambda_3$ seed=42"  ]
axis_test_label_seed84   = ["$\lambda_1$ seed=84"  ,"$\lambda_2$ seed=84"  ,"$\lambda_3$ seed=84"  ]
axis_test_label_seed167   = ["$\lambda_1$ seed=167"  ,"$\lambda_2$ seed=167"  ,"$\lambda_3$ seed=167"  ]
axs[0].set_title("Relative Errors (True LEs vs RC LEs, N = " + str(N) + ")")
                 
for j in range(2):
    if j==0:
        i=0
    if j==1:
        i=2
    axs[j].plot(rho_list, df_RC_LE_seed42.iloc[:, [i+8]] , label= axis_test_label_seed42[i], color='red' , lw = 0.5)
    axs[j].plot(rho_list, df_RC_LE_seed84.iloc[:, [i+8]] , label= axis_test_label_seed84[i], color='tab:green' , lw = 0.5)
    axs[j].plot(rho_list, df_RC_LE_seed167.iloc[:, [i+8]] , label= axis_test_label_seed167[i], color='tab:blue' , lw = 0.5)

    #
    # Shrink current axis by 20%
    box = axs[j].get_position()
    axs[j].set_position([box.x0, box.y0, box.width * 0.8, box.height])
    # Put a legend to the right of the current axis
    axs[j].legend(loc='center left', bbox_to_anchor=(1, 0.5))

    axs[j].axvline(x = 13.926667, color = 'grey')
    axs[j].axvline(x = 23.9     , color = 'grey')
    #axs[j].axvline(x = 24.058   , color = 'grey')
    #axs[j].axvline(x = 470./19. , color = 'grey')
    #axs[j].axvline(x = 30.485   , color = 'grey')

    axs[j].axvline(x = 75     , color = 'grey')

    #axs[j].axvline(x = 99.524   , color = 'grey')
    #axs[j].axvline(x = 100.795  , color = 'grey')
    #axs[j].axvline(x = 148.4    , color = 'grey')
    axs[j].axvline(x = 166.07   , color = 'grey')
    #axs[j].axvline(x = 214.364  , color = 'grey')
    #axs[j].axvline(x = 233.5    , color = 'grey')
    axs[j].axvline(x = 313      , color = 'grey')
    #axs[j].axvline(x = 75, color = 'grey')
    axs[j].set_xlim(-5, 335)
    axs[j].set_ylim(0, 4)

    axs[j].set_yticks([0.38,4])

    axs[j].axhline(y = 0.38 , color = 'grey')

    if i==0:
        ax2 = axs[i].twiny()
        ax2.set_xlim(-5, 335)
        ax2.set_xticks([13.926667,23.9,75,166.07, 313])
        #ax2.set_xticklabels(['$\\rho_H$','$\\rho_a =23.9$','166.07','313'],color='purple',fontsize=8,ha='left')
        ax2.set_xticklabels(['$\\rho_H$','$\\rho_a$','75','166.07','313'],color='purple')
    
    
axs[1].set_xlabel("$\\rho$") #\\rho need to prevent parsing error!!!
fig.tight_layout()

plt.savefig('climate/'+ 'REL_ERR_LE_N' + str(N) + '.png',bbox_inches='tight')

plt.show()

plt.close(fig)
```
